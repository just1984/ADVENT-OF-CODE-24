# Task 1

```python
import pandas as pd
from collections import Counter

file_path = "2412_01_input.txt"
data = pd.read_csv(file_path, sep=r"\s+", header=None)

left_list = sorted(data[0])
right_list = sorted(data[1])

left_list, right_list
```

```python
differences = [abs(left - right) for left, right in zip(left_list, right_list)]

total_difference = sum(differences)

print("Differenzen:", differences)
print(f"Gesamtsumme der Differenzen: {total_difference}")

total_difference
```

# Task 2

```python
right_count = Counter(right_list)

similarity_score = sum(left * right_count[left] for left in left_list)

print(f"HÃ¤ufigkeiten der rechten Liste: {right_count}")
print(f"Similarity Score: {similarity_score}")

similarity_score
```
