```python
content = open('6_12.csv').readlines()
```

```python
total =[]

for line in content:
    l = list(line[:-1])
    total.append(l)
```

```python
def walk(total):
    result =''
    for index in range(len(total)):
        for i in range(len(total[index])):

            if total[index][i]=='^':

                    for k in range(len(total)):
                        if (index-k)>0:
                            if total[index-1-k][i]!='#':
                                total[index-1-k][i]='X'
                            else:
                                total[index-k][i]='>'
                                break
                        else:
                            result ='Done'
                            break
            if total[index][i]=='>':

                    for k in range(len(total)):
                        if (i+k)<129:
                            if total[index][i+1+k]!='#':
                                total[index][i+1+k]='X'
                            else:
                                total[index][i+k]='v'
                                break
                        else:
                            result ='Done'
                            break
            if total[index][i]=='v':

                    for k in range(len(total)):
                        if (index+k)<129:
                            if total[index+1+k][i]!='#':
                                total[index+1+k][i]='X'
                            else:
                                total[index+k][i]='<'
                                break
                    else:
                        result ='Done'
                        break
            if total[index][i]=='<':

                    for k in range(len(total)):
                        if (i-k)>0:
                            if total[index][i-1-k]!='#':
                                total[index][i-1-k]='X'
                            else:
                                total[index][i-k]='^'
                                break
                    else:
                        result = 'Done'
                        break
    return result
```

```python
i=0
#walk till end of the map
while walk(total)!='Done':
    i+=1
    walk(total)
```

```python
for t in total:
    print(" ".join(map(str, t)))
```

```python
#ground covered
sum=0

for line in total:
    for l in line:

        if l =='X' or l =='v'or l =='<'or l =='^'or l =='>':
            sum+=1
print(f'Answer for Part 1: {sum}')
```

```python
import copy
def add_obstacle(data,line,column):
    temp_total = copy.deepcopy(data)

    temp_total[line][column]='#'

    return temp_total
```

```python
count =0
answer = 0
for l in range(0,129):
    for c in range(0,129):
        count+=1
        print(f'count:{count}')
        x=0

        temp_total = add_obstacle(total,l,c)
        while walk(temp_total)!='Done' and x<=100:
            x+=1

            walk(temp_total)
        if x==101:
            answer+=1
            print(f'answer:{answer}')
```
